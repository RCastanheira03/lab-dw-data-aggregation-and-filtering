```python
import pandas as pd

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"

data = pd.read_csv(url)
```

```python
data.head()
```

```python
null_columns = data.isnull().any()
print(null_columns)
```

```python
null_columns = ["Response", "Months Since Last Claim", "Number of Open Complaints", "Vehicle Class", "Vehicle Size", "Vehicle Type"]

data[null_columns] = data[null_columns].fillna("Unknown")
```

```python
data.duplicated().any()
```

```python
data["Customer Lifetime Value"] = data["Customer Lifetime Value"].round(1)
data["Total Claim Amount"] = data["Customer Lifetime Value"].round(1)
```

```python
data.columns
```

```python
# Create a new DataFrame that only includes customers who:
# have a low total_claim_amount (e.g., below $1,000),
# have a response "Yes" to the last marketing campaign.

same_columns = ["Total Claim Amount", "Response"]

new_data = data[same_columns].copy()

new_data = new_data[(new_data["Total Claim Amount"] < 3000) & (new_data["Response"] == "Yes")]
new_data
```

```python
# Using the original Dataframe, analyze:
# the average monthly_premium and/or customer lifetime value by policy_type and gender for customers who responded "Yes"

to_analyze = ["Customer Lifetime Value", "Monthly Premium Auto", "Response", "Gender", "Policy Type"]

new_data2 = data[to_analyze].copy()
new_data2
```

```python
new_data2_yes = new_data2[new_data2["Response"] == "Yes"]

result = (new_data2_yes.groupby(["Policy Type", "Gender"])[["Monthly Premium Auto", "Customer Lifetime Value"]].mean().reset_index())
display(result)
```

```python
# compare these insights to total_claim_amount patterns, and discuss which segments appear most profitable or low-risk for the company.

comparison = ["Total Claim Amount", "Response", "Gender", "Policy Type"]

data_comparison = data[comparison].copy()
comparison_yes = data_comparison[data_comparison["Response"] == "Yes"]
comparison_yes.describe() # When comparing results, we can see the average claim amount for clients that match the criteria is 7882.
# From this information, the most profitable policy type is Special Auto but it is almost identical to Personal Auto.
```

```python
# Analyze the total number of customers who have policies in each state, 
# and then filter the results to only include states where there are more than 500 customers.

state_data = ["State"]
new_state_data = data[state_data].copy()

state_counts = new_state_data["State"].value_counts()
print(state_counts)

#All states have over 500 customers
```

```python
# Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

for_statistics = ["Customer Lifetime Value", "Education", "Gender"]

statistics_data = data[for_statistics].copy()
statistics_data

summary = (
    statistics_data.groupby(["Education", "Gender"])["Customer Lifetime Value"]
    .agg(["min", "max", "median"])
    .reset_index()
)

display(summary)
```
